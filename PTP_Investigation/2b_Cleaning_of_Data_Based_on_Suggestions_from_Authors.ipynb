```python
import pandas as pd
import numpy as np
```

```python
feat_y = pd.read_csv('./features/palma_2017_experimental_results_and_sec_scores.csv', index_col=0)
```

```python
found = pd.merge(feat_y, feat_x, on='uid_pos')
```

```python
feat_y_r = feat_y[feat_y['uid_pos'].isin(found['uid_pos']) == False]
feat_x_r = feat_x[feat_x['uid_pos'].isin(found['uid_pos']) == False]
```

```python
temp = feat_x_r['uid_pos'].str.split('_', expand=True)
```

```python
temp[1] = temp[1].astype(int) + 7
```

```python
feat_x_r['uid_pos'] = temp[0] + '_' + temp[1].astype(str)
```

```python
feat_x_r.loc[feat_x_r['uid_pos'] == 'Q9BTX7_70', 'uid_pos'] = 'Q9BTX7_69'
feat_x_r.loc[feat_x_r['uid_pos'] == 'Q53H88_7', 'uid_pos'] = 'Q53H88_6'
feat_x_r.loc[feat_x_r['uid_pos'] == 'P31268_7', 'uid_pos'] = 'P31268_6'
```

```python
full = pd.merge(feat_x_r, feat_y_r, on='uid_pos', how='left')
```

```python
final_feat_x = final[feat_x.columns.drop('uid_pos')]
```

```python
final_feat_y = final[feat_y.columns.drop('uid_pos')]
```

```python
final_feat_y = final_feat_y.rename(columns={'PTP_HD-PTP_11-07-2008.seam':'PTP_HD-PTP', 'PTP_LAR_12-10-2007.seam':'PTP_LAR',
       'PTP_LyP_12-10-2007.seam':'PTP_LyP', 'PTP_MEG-1_12-10-2007.seam':'PTP_MEG-1',
       'PTP_MEG-2_12-10-2007.seam':'PTP_MEG-2', 'PTP_PTP-PEST_12-10-2007.seam':'PTP_PTP-PEST',
       'PTP_PTPH1_12-10-2007.seam':'PTP_PTPH1', 'PTP_rPTP-alpha_12-10-2007.seam':'PTP_rPTP-alpha',
       'PTP_rPTP-beta_12-10-2007.seam':'PTP_rPTP-beta', 'PTP_SAP-1_12-10-2007.seam':'PTP_SAP-1',
       'PTP_SHP-1_12-10-2007.seam':'PTP_SHP-1', 'PTP_SHP-2_12-10-2007.seam':'PTP_SHP-2',
       'PTP_DEP-1_12-10-2007.seam.txt':'PTP_DEP-1', 'PTP_TC-PTP_12-10-2007.seam':'PTP_TC-PTP'})
```

```python
PTP_cols = ['PTP_HD-PTP', 'PTP_LAR', 'PTP_LyP', 'PTP_MEG-1', 'PTP_MEG-2',
       'PTP_PTP-PEST', 'PTP_PTPH1', 'PTP_rPTP-alpha', 'PTP_rPTP-beta',
       'PTP_SAP-1', 'PTP_SHP-1', 'PTP_SHP-2', 'PTP_DEP-1', 'PTP_TC-PTP',
       'PTP_PTP1B_averaged']
```

```python
p = 0
while p < len(PTP_cols):
    pos = len(final_feat_y[final_feat_y[PTP_cols[p]]  >= 1.85])
    print('within', PTP_cols[p], pos, 'of', len(final_feat_y), 'are positive with a 1.85 cutoff (', pos/len(final_feat_y)*100, '%)')
    p+=1
```

```python
## THE MAIN TEXT WILL INCLUDE DEP-1 AND PTP1B AS WE KNOW THE CUTOFFS; APPLY
## STATIC CUTOFF FOR THE OTHER 13 ENZYMES
```

```python
## From Dr. Tinti -- drop the peptides with Spot Flags that aren't "GOOD"
good_y = final_feat_y[final_feat_y['Spot Flag'] == 'GOOD']
good_x = final_feat_x.loc[good_y.index]

# Unfortunately this reduces our dataset by 2,073 but we can be sure our readings
#  are more reliable! (New total length of dataset = 3,984)
```

```python
## NOW ONTO GENERATING OUR BINARY LABELS
# PTP1B -- take the top 300 entries as positive, the rest are negative (lowest score = 1.7)
good_y = good_y.sort_values(by='PTP_PTP1B_averaged', ascending=False)

good_y['PTP1B_Binary'] = 0
good_y.loc[good_y.head(n=300).index, 'PTP1B_Binary'] = 1
```

```python
# DEP-1 -- take the top 215 entries as positive, the rest are negative (lowest score = 1.79)
good_y = good_y.sort_values(by='PTP_DEP-1', ascending=False)

good_y['DEP1_Binary'] = 0
good_y.loc[good_y.head(n=215).index, 'DEP1_Binary'] = 1
```

```python
# Considering the lowest cutoff applied for readings is 1.7, and the same microarray
#  chip was used for all experiments, we will assume that a cutoff of 1.7 can be applied
#  to the remaining datasets to discern positive/negative data points
# NOTE: These 13 investigations will be only presented in the supplementary information
#  as they rely heavily on assumptions and may in reality be including false positive
#  or negative points. 
```

```python
remaining = ['PTP_HD-PTP', 'PTP_LAR', 'PTP_LyP', 'PTP_MEG-1', 'PTP_MEG-2',
       'PTP_PTP-PEST', 'PTP_PTPH1', 'PTP_rPTP-alpha', 'PTP_rPTP-beta',
       'PTP_SAP-1', 'PTP_SHP-1', 'PTP_SHP-2', 'PTP_TC-PTP']
```

```python
test_gy = good_y
```

```python
r = 0
while r < len(remaining):
    # create column for binary results to go into
    test_gy[remaining[r][4:]+'_Binary'] = 0
    # if bigger than 1.7, mark as positive
    test_gy.loc[test_gy[remaining[r]] > 1.7, remaining[r][4:]+'_Binary'] = 1
    print(len(test_gy[test_gy[remaining[r]] > 1.7]), 'larger than 1.7 within', remaining[r])
    print('sum of', test_gy[remaining[r][4:]+'_Binary'].sum(), 'in the', (remaining[r-1][4:]+'_Binary'), 'column\n')
    r+=1
```
